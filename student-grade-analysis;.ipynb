```python
import pandas as pd 

# Goals: 
# - read csv 
# - find average score
# - find highest and lowest marks
# - fliter students with score > 80
# - add total and percentage columns
# - sort students by percentage

df = pd.read_csv("StudentsPerformance.csv")

## average score for each subjects math, reading, writting 
math_average = df['math score'].mean()
print('math average score is: ', +math_average)

reading_average = df['reading score'].mean()
print('reading average score is: ',reading_average)

writting_score = df['writing score'].mean()
print('writting average score is: ', writting_score)

# ------------------------------------------------------------------------- #
highest_math_score = df['math score'].max()
print('highest math score', highest_math_score)

highest_reading_score = df['reading score'].max()
print('highest reading score', highest_math_score)

highest_writing_score = df['writing score'].max()
print('highest writing score', highest_writing_score)

# for lowest
lowest_math_score = df['math score'].min()
print('lowest math score', lowest_math_score)

lowest_reading_score = df['reading score'].min()
print('lowest reading score', lowest_math_score)

lowest_writing_score = df['writing score'].min()
print('lowest writing score', lowest_writing_score)

# ------------------------------------------------------------------------- #

students_with_higer_than_80_marks = df[df['math score'] > 80]
print(students_with_higer_than_80_marks)

# ------------------------------------------------------------------------- #

# Add a Total column

# df['total'] = '' or None or 0 -- to create an empty column

df['Total'] = (df["math score"] + df["reading score"] + df["writing score"])

df['Percentage'] = ((df['Total'] / 300) * 100 )

# print(df.head())

# ------------------------------------------------------------------------- #

# sort student by percentage

df = df.sort_values(by='Percentage', ascending=False)


# ------------------------------------------------------------------------- #

df.head()

# print(df)


```
